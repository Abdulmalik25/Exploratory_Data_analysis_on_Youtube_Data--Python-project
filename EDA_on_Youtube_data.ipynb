**5)Convert datatype of column published_at from object to pandas datetime.**

```python
data['published_at']=pd.to_datetime(data['published_at'])
data
```

**6) Create a new column as 'published_month' using the column published_at (display the months only).**

```python
data['published_month']=data['published_at'].dt.month
data
```

**7) Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'
and so on.....**

```python
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr',
    5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug',
    9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
data['published_month']=data['published_month'].map(month_names)
data
```

**8) Find the number of videos published each month and arrange the months in a decreasing order based

**13) Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.**

```python
correlation_coefficient=data['view_count'].corr(data['dislikes'])
print('Correlation coffecient between view_count and dislikes is :',correlation_coefficient)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=data,x='view_count', y='dislikes', color='blue')
plt.title('Scatter Plot of Views vs Dislikes')
plt.xlabel('Number of Views')
plt.ylabel('Number of Dislikes')
plt.show()
```

**Observation**- Correlation between view counts and number of dislikes is positive which is around (0.68). The scatterplot is likely to show a general upward trend, indicating that videos with higher view counts tend to have a higher number of dislikes.
